### Constant n and p BEGE
The first model that I estimated is BEGE with invariant shape parameters, which is 

\begin{align}
\pi_{t+1} &= Mean_{t} +u_{t+1}\\
u_{t+1} &= \sigma_p \omega_{p} - \sigma_n \omega_{n}
\end{align}
where
\begin{equation}
\omega_{p} \sim \tilde{\Gamma}( \bar{p}, 1), \quad \omega_{n} \sim \tilde{\Gamma}(\bar{n}, 1).
\end{equation}
$ \bar{p}$ and $\bar{n}$ are unconditional shape parameters. 
 

$
\pi (t) =  fc(t-1) + \epsilon (t)

```python
# Jupyter cell: summarize draw_*.pkl into per-folder CSVs with AIC, BIC, loglik, params

import re
import pickle
from pathlib import Path
from typing import Any, Dict, List, Optional

import numpy as np
import pandas as pd

# =======================
# Configuration
# =======================
# Parent directory whose immediate subfolders contain draw_*.pkl files
PARENT_DIR = Path("/project/lhansen/Capital_NN_variant/BEGE_GARCH/RandomDraw_Symmetric_New")

# Name of the CSV to write inside each subfolder
OUTPUT_CSV_NAME = "summary_draws.csv"

# Filename pattern: draw_1.pkl, draw_200.pkl, ...
DRAW_FILE_REGEX = re.compile(r"^draw_(\d+)\.pkl$")


# =======================
# Helpers
# =======================
def find_draw_files(folder: Path) -> List[Path]:
    """Return draw_*.pkl files in a folder, sorted by numeric ID."""
    items = []
    for p in folder.iterdir():
        if p.is_file() and p.suffix == ".pkl":
            m = DRAW_FILE_REGEX.match(p.name)
            if m:
                items.append((int(m.group(1)), p))
    items.sort(key=lambda x: x[0])
    return [p for _, p in items]

def safe_load_pickle(pkl_path: Path):
    """Load pickle (trusted environment assumed)."""
    with pkl_path.open("rb") as f:
        return pickle.load(f)

def to_float_scalar(x: Any) -> float:
    """
    Convert numpy scalars or arrays of shape () to float.
    If x is None or not scalar-coercible, return NaN.
    """
    if x is None:
        return float("nan")
    try:
        arr = np.asarray(x)
        if arr.shape == ():
            return float(arr)
        # Try item() if it behaves like a 0-d array
        return float(arr.item())
    except Exception:
        try:
            return float(x)
        except Exception:
            return float("nan")

def extract_records_from_obj(obj: Any, draw_id: Optional[int]) -> List[Dict[str, Any]]:
    """
    From a loaded pickle object (dict or list of dicts), extract rows with AIC, BIC, loglik, params.
    If 'obj' is a list, produce one row per element, adding rep_idx = 0,1,2,... .
    """
    def _extract_one(d: Dict[str, Any], rep_idx: Optional[int]) -> Dict[str, Any]:
        AIC = to_float_scalar(d.get("AIC", None))
        BIC = to_float_scalar(d.get("BIC", None))
        loglik = to_float_scalar(d.get("loglik", None))
        params = d.get("params", None)
        params = np.asarray(params).ravel() if params is not None else np.array([], dtype=float)

        return {
            "draw_id": draw_id,
            "rep_idx": rep_idx,     # None if single dict; integer if from a list
            "AIC": AIC,
            "BIC": BIC,
            "loglik": loglik,
            "params": params,
        }

    records = []
    if isinstance(obj, dict):
        records.append(_extract_one(obj, rep_idx=None))
    elif isinstance(obj, list):
        for i, elem in enumerate(obj):
            if isinstance(elem, dict):
                records.append(_extract_one(elem, rep_idx=i))
            # silently skip unexpected elements
    # silently skip unknown types
    return records

def records_to_df(records: List[Dict[str, Any]]) -> pd.DataFrame:
    """Flatten 'params' arrays into columns param_1..param_k and return a DataFrame."""
    if not records:
        return pd.DataFrame(columns=["draw_id", "rep_idx", "AIC", "BIC", "loglik"])

    max_len = max(len(r["params"]) for r in records)
    rows = []
    for r in records:
        base = {
            "draw_id": r["draw_id"],
            "rep_idx": r["rep_idx"],
            "AIC": r["AIC"],
            "BIC": r["BIC"],
            "loglik": r["loglik"],
        }
        p = r["params"]
        for i in range(max_len):
            base[f"param_{i+1}"] = float(p[i]) if i < len(p) else np.nan
        rows.append(base)

    df = pd.DataFrame(rows)

    # sort nicely: by draw_id then rep_idx (if present)
    sort_cols = ["draw_id"]
    if df["rep_idx"].notna().any():
        sort_cols.append("rep_idx")
    df = df.sort_values(sort_cols).reset_index(drop=True)
    return df

def process_subfolder(folder: Path, output_csv_name: str = OUTPUT_CSV_NAME) -> Optional[Path]:
    draw_files = find_draw_files(folder)
    if not draw_files:
        print(f"[SKIP] {folder} has no draw_*.pkl files.")
        return None

    all_records: List[Dict[str, Any]] = []
    errors = []

    for p in draw_files:
        m = DRAW_FILE_REGEX.match(p.name)
        draw_id = int(m.group(1)) if m else None
        try:
            obj = safe_load_pickle(p)
            recs = extract_records_from_obj(obj, draw_id=draw_id)
            all_records.extend(recs)
        except Exception as e:
            errors.append((p.name, str(e)))

    df = records_to_df(all_records)
    out_path = folder / output_csv_name
    df.to_csv(out_path, index=False)

    print(f"[OK] {folder.name}: wrote {len(df)} rows -> {out_path}")
    if errors:
        print(f"[WARN] {folder.name}: {len(errors)} file(s) failed to parse:")
        for name, msg in errors[:10]:
            print(f"   - {name}: {msg}")
        if len(errors) > 10:
            print(f"   ... and {len(errors)-10} more")

    return out_path


# =======================
# Run over all subfolders
# =======================
parent = PARENT_DIR
if not parent.exists():
    raise FileNotFoundError(f"{parent} does not exist!")

subfolders = [d for d in parent.iterdir() if d.is_dir()]
if not subfolders:
    print(f"[INFO] No subfolders found under {parent}")
else:
    outputs = []
    for sf in sorted(subfolders):
        out = process_subfolder(sf)
        if out:
            outputs.append(out)

    print("\nDone. Generated:")
    for p in outputs:
        print(" -", p)

    # Optional: peek at the first CSV
    if outputs:
        display(pd.read_csv(outputs[0]).head())
```

```python
# Jupyter cell: summarize draw_*.pkl into per-folder CSVs with AIC, BIC, loglik, params (named)

import re
import pickle
from pathlib import Path
from typing import Any, Dict, List, Optional

import numpy as np
import pandas as pd

# =======================
# Configuration
# =======================
# Parent directory whose immediate subfolders contain draw_*.pkl files
PARENT_DIR = Path("/project/lhansen/Capital_NN_variant/BEGE_GARCH/RandomDraw_BG_GARCH")

# PARENT_DIR = Path("/project/lhansen/Capital_NN_variant/BEGE_GARCH/RandomDraw_ID")
# PARENT_DIR = Path("/project/lhansen/Capital_NN_variant/BEGE_GARCH/RandomDraw_Full_Dec")
# Name of the CSV to write inside each subfolder
OUTPUT_CSV_NAME = "summary_draws.csv"

# Filename pattern: draw_1.pkl, draw_200.pkl, ...
DRAW_FILE_REGEX = re.compile(r"^draw_(\d+)\.pkl$")


# =======================
# Helpers
# =======================
def find_draw_files(folder: Path) -> List[Path]:
    """Return draw_*.pkl files in a folder, sorted by numeric ID."""
    items = []
    for p in folder.iterdir():
        if p.is_file() and p.suffix == ".pkl":
            m = DRAW_FILE_REGEX.match(p.name)
            if m:
                items.append((int(m.group(1)), p))
    items.sort(key=lambda x: x[0])
    return [p for _, p in items]

def safe_load_pickle(pkl_path: Path):
    """Load pickle (trusted environment assumed)."""
    with pkl_path.open("rb") as f:
        return pickle.load(f)

def to_float_scalar(x: Any) -> float:
    """
    Convert numpy scalars or arrays of shape () to float.
    If x is None or not scalar-coercible, return NaN.
    """
    if x is None:
        return float("nan")
    try:
        arr = np.asarray(x)
        if arr.shape == ():
            return float(arr)
        # Try item() if it behaves like a 0-d array
        return float(arr.item())
    except Exception:
        try:
            return float(x)
        except Exception:
            return float("nan")

def extract_records_from_obj(obj: Any, draw_id: Optional[int]) -> List[Dict[str, Any]]:
    """
    From a loaded pickle object (dict or list of dicts), extract rows with AIC, BIC, loglik, params, names.
    If 'obj' is a list, produce one row per element, adding rep_idx = 0,1,2,... .
    Each record carries 'param_names' (list[str] or [] if unavailable).
    """
    def _extract_one(d: Dict[str, Any], rep_idx: Optional[int]) -> Dict[str, Any]:
        AIC = to_float_scalar(d.get("AIC", None))
        BIC = to_float_scalar(d.get("BIC", None))
        loglik = to_float_scalar(d.get("loglik", None))
        params = d.get("params", None)
        params = np.asarray(params).ravel() if params is not None else np.array([], dtype=float)
        # Pull names if present
        names = d.get("names", None)
        if names is None:
            param_names: List[str] = []
        else:
            # Ensure a flat list of strings
            try:
                param_names = [str(x) for x in list(names)]
            except Exception:
                param_names = []
        return {
            "draw_id": draw_id,
            "rep_idx": rep_idx,     # None if single dict; integer if from a list
            "AIC": AIC,
            "BIC": BIC,
            "loglik": loglik,
            "params": params,
            "param_names": param_names
        }

    records: List[Dict[str, Any]] = []
    if isinstance(obj, dict):
        records.append(_extract_one(obj, rep_idx=None))
    elif isinstance(obj, list):
        for i, elem in enumerate(obj):
            if isinstance(elem, dict):
                records.append(_extract_one(elem, rep_idx=i))
            # silently skip unexpected elements
    # silently skip unknown types
    return records

def _choose_param_headers(records: List[Dict[str, Any]]) -> List[str]:
    """
    Decide parameter column headers.
    If all records that have names provide the same-length list, use the first such list.
    Otherwise, fall back to param_1..param_k where k = max length across records.
    """
    if not records:
        return []

    # Maximum parameter length
    max_len = max(len(r["params"]) for r in records)

    # Gather candidate name lists that match their params length and are non-empty
    candidates: List[List[str]] = []
    for r in records:
        names = r.get("param_names", [])
        if names and len(names) == len(r["params"]):
            candidates.append(names)

    # If we have at least one valid names list, check if lengths align with max_len
    # and use the first candidate that matches max_len; else consider the first candidate anyway.
    chosen: Optional[List[str]] = None
    for cand in candidates:
        if len(cand) == max_len:
            chosen = cand
            break
    if chosen is None and candidates:
        chosen = candidates[0]  # fallback: first names list, even if shorter

    # Normalize to max_len (pad with generic names if needed)
    if chosen is None:
        # No valid names anywhere -> generic names
        return [f"param_{i+1}" for i in range(max_len)]

    headers = list(chosen[:max_len])
    if len(headers) < max_len:
        # pad with generic names for missing tails
        headers += [f"param_{i+1}" for i in range(len(headers), max_len)]
    # Ensure uniqueness (avoid duplicates by appending suffixes if necessary)
    seen = {}
    unique_headers = []
    for h in headers:
        if h not in seen:
            seen[h] = 1
            unique_headers.append(h)
        else:
            seen[h] += 1
            unique_headers.append(f"{h}__{seen[h]}")
    return unique_headers

def records_to_df(records: List[Dict[str, Any]]) -> pd.DataFrame:
    """Flatten 'params' arrays into columns using named headers when available."""
    if not records:
        return pd.DataFrame(columns=["draw_id", "rep_idx", "AIC", "BIC", "loglik"])

    # Decide headers
    param_headers = _choose_param_headers(records)
    max_len = len(param_headers)

    rows = []
    for r in records:
        base = {
            "draw_id": r["draw_id"],
            "rep_idx": r["rep_idx"],
            "AIC": r["AIC"],
            "BIC": r["BIC"],
            "loglik": r["loglik"],
        }
        p = r["params"]
        # Write params by header order
        for i, col in enumerate(param_headers):
            base[col] = float(p[i]) if i < len(p) else np.nan
        rows.append(base)

    df = pd.DataFrame(rows)

    # sort nicely: by draw_id then rep_idx (if present)
    sort_cols = ["draw_id"]
    if df["rep_idx"].notna().any():
        sort_cols.append("rep_idx")
    df = df.sort_values(sort_cols).reset_index(drop=True)
    return df

def process_subfolder(folder: Path, output_csv_name: str = OUTPUT_CSV_NAME) -> Optional[Path]:
    draw_files = find_draw_files(folder)
    if not draw_files:
        print(f("[SKIP] {folder} has no draw_*.pkl files."))
        return None

    all_records: List[Dict[str, Any]] = []
    errors = []

    for p in draw_files:
        m = DRAW_FILE_REGEX.match(p.name)
        draw_id = int(m.group(1)) if m else None
        try:
            obj = safe_load_pickle(p)
            recs = extract_records_from_obj(obj, draw_id=draw_id)
            all_records.extend(recs)
        except Exception as e:
            errors.append((p.name, str(e)))

    df = records_to_df(all_records)
    out_path = folder / output_csv_name
    df.to_csv(out_path, index=False)

    print(f"[OK] {folder.name}: wrote {len(df)} rows -> {out_path}")
    if errors:
        print(f"[WARN] {folder.name}: {len(errors)} file(s) failed to parse:")
        for name, msg in errors[:10]:
            print(f"   - {name}: {msg}")
        if len(errors) > 10:
            print(f"   ... and {len(errors)-10} more")

    return out_path


# =======================
# Run over all subfolders
# =======================
parent = PARENT_DIR
if not parent.exists():
    raise FileNotFoundError(f"{parent} does not exist!")

subfolders = [d for d in parent.iterdir() if d.is_dir()]
if not subfolders:
    print(f"[INFO] No subfolders found under {parent}")
else:
    outputs = []
    for sf in sorted(subfolders):
        out = process_subfolder(sf)
        if out:
            outputs.append(out)

    print("\nDone. Generated:")
    for p in outputs:
        print(" -", p)

    # Optional: peek at the first CSV
    if outputs:
        display(pd.read_csv(outputs[0]).head())
```
